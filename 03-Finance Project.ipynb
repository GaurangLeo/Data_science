** Figure out how to get the stock data from Jan 1st 2006 to Jan 1st 2025 for each of these banks. Set each bank to be a separate dataframe, with the variable name for that bank being its ticker symbol. This will involve a few steps:**
1. Use datetime to set start and end datetime objects.

```python
start = datetime.datetime(2006, 1, 1)
end = datetime.datetime(2016, 1, 1)
```

```python
import yfinance as yf

# Override pandas_datareader's default method to use yfinance
#yf.pdr_override()


# Fetch stock data
#Bank of America
BAC = yf.download("BAC", start=start, end=end)
# Citibank
C = yf.download("C", start=start, end=end)
# Goldman sachs
GS = yf.download("GS", start=start, end=end)
#  JPMorgan Chase
JPM = yf.download("JPM", start=start, end=end)
# Morgan stanley
MS = yf.download("MS", start=start, end=end)
# Wells fargo
WFC = yf.download("WFC", start=start, end=end)
```

** Create a list of the ticker symbols (as strings) in alphabetical order. Call this list: tickers**

```python
tickers = ['BAC', 'C', 'GS','JPM', 'MS','WFC']
```

** Create a new empty DataFrame called returns. This dataframe will contain the returns for each bank's stock. returns are typically defined by:**

$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

```python
returns = pd.DataFrame(index=bank_stocks.index)
```

** We can use pandas pct_change() method on the Close column to create a column representing this return value. Create a for loop that goes and for each Bank Stock Ticker creates this returns column and set's it as a column in the returns DataFrame.**

```python
for tick in tickers:
    returns[tick + ' Return'] = bank_stocks['Close',tick].pct_change()

returns.head()
```

## Moving Averages

Let's analyze the moving averages for these stocks in the year 2008. 

** Plot the rolling 30 day average against the Close Price for Bank Of America's stock for the year 2008**

```python
BAC
```

```python
BAC.index = pd.to_datetime(BAC.index)
BAC
```

```python
data = BAC['Close'].loc['2008-01-01':'2009-01-01']
rolling_mean = data.rolling(window=30).mean()

fig, ax = plt.subplots(figsize=(12,4))

# Plot with no label to avoid pandas auto-labeling confusion
rolling_mean.plot(ax=ax, color='orange', label='_nolegend_')  
data.plot(ax=ax, color='blue', label='_nolegend_')

# Manually create legend handles and labels
line1, = ax.plot([], [], color='orange', label='30 day Mov avg')
line2, = ax.plot([], [], color='blue', label='BAC Close')

# Add the manual legend
ax.legend(handles=[line1, line2])

ax.set_xlabel('Date')
ax.set_ylabel('Price')
ax.set_title('BAC Close Price and 30-day Moving Average (2008)')
plt.show()
```

** Create a heatmap of the correlation between the stocks Close Price.**

```python
corr_matrix = bank_stocks.xs(key='Close', axis=1, level='Price').corr()
print(corr_matrix)
print(corr_matrix.dtypes)
```

```python
print(corr_matrix.isnull().sum())
```

```python
plt.figure(figsize=(6,6))
sns.heatmap(corr_matrix, cmap='seismic', fmt=".4f", annot_kws={"color": "black"})
plt.title('Correlation of Bank Closing Prices')
plt.tight_layout()
plt.show()

```

** Optional: Use seaborn's clustermap to cluster the correlations together:**

```python
plt.figure(figsize=(6,6))
sns.clustermap(corr_matrix, cmap='seismic', fmt=".4f", annot_kws={"color": "black"})
plt.title('Correlation of Bank Closing Prices')
plt.tight_layout()
plt.show()
```
